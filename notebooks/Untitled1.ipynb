```python
import numpy as np
from sklearn.datasets import make_blobs, make_moons
from sklearn.ensemble import RandomForestClassifier
import plotly.graph_objs as go
import plotly_express as px

from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
data_type = "blobs"

if data_type == "blobs":
    x_train, y_train = make_blobs(centers=2, n_samples=300, n_features=2, cluster_std=0.05 * 47 + 0.57)
    x_test, y_test = make_blobs(centers=2, n_samples=150, n_features=2, cluster_std=0.05 * 47 + 0.57)
    
    scaler = StandardScaler()
    x_train = scaler.fit_transform(x_train)
    x_test = scaler.transform(x_test)
    
elif data_type == "moons":
    x_train, y_train = make_moons(n_samples=300, noise=0.05)
    x_test, y_test = make_moons(n_samples=150, noise=0.05)
```

```python
model = RandomForestClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
def plot_decision_boundary(model, x_train, y_train, x_test, y_test):
    x_min, x_max = x_train[:, 0].min() - 1, x_train[:, 0].max() + 1
    y_min, y_max = x_train[:, 1].min() - 1, x_train[:, 1].max() + 1

    h = 0.02

    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    y_ = np.arange(y_min, y_max, h)

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    fig = go.Figure(
        data=[go.Heatmap(x=xx[0], y=y_, z=Z, colorscale="oxy", showscale=False)],
    )

    trace1 = go.Scatter(
        x=x_train[:, 0],
        y=x_train[:, 1],
        name="train data",
        mode="markers",
        showlegend=True,
        marker=dict(
            size=10, color=y_train, colorscale="oxy", line=dict(color="white", width=2)
        ),
    )

    fig.add_trace(trace1)

    trace2 = go.Scatter(
        x=x_test[:, 0],
        y=x_test[:, 1],
        name="test data",
        mode="markers",
        showlegend=True,
        marker_symbol="cross",
        visible="legendonly",
        marker=dict(
            size=10,
            color=y_test,
            colorscale="oxy",
            line=dict(color="white", width=2),
        ),
    )
    fig.add_trace(trace2)

    fig.update_layout(height=450)
    fig.update_xaxes(range=[x_min, x_max], title="x1")
    fig.update_yaxes(range=[y_min, y_max], title="x2")

    return fig
```

```python
%%time
fig = plot_decision_boundary(model, x_train, y_train, x_test, y_test)
```
