```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_classification
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
```

```python
def plot_data(x1, x2, y=None):
    plt.scatter(x1, x2, c=y)
```

```python
x, y = make_classification(n_samples=1000, n_features=2, n_redundant=0, n_repeated=0, class_sep=2)


plot_data(x[:, 0], x[:, 1], y)
```

```python
model = GaussianNB()
```

```python
model.fit(x, y)
```

```python
metrics.accuracy_score(y, model.predict(x))
```
